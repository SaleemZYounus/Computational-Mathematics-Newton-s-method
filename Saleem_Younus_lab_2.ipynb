## Secant Method

### Exercise 9

#### Part 1
Write a function that take into argument a function, its derivative, an initial guess, a tolerance, a maximum number of iteration and return the zero of the function using Secant's method


```python

def my_secant(f, df, x0, x1, tol, itmax):
  step = 0
  

  while step < itmax:
    secant = x1 - f(x1)/( (f(x1) - f(x0))/(x1 - x0) )
    x0 = secant
   

    step += 1
    if abs(f(x0)) < tol:
      break
    else:
      None #return my_secent(f, df, x0 - f(x0)/df(x0), tol)
  print(f"The root approx is: {x0}, found with {step} iterations. {np.abs(f(x0))} ≈ 0")
  x = np.linspace(-10,10,100)
  plt.plot(x, f(x),label='f(x)')
  plt.legend()
  plt.grid()
  plt.ylim((-20,20))
  plt.show()
  return x0, step
```
